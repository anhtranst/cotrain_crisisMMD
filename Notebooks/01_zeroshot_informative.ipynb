# 01 — Zero-Shot Informative Classification (Llama-3.2-11B-Vision-Instruct)

Run zero-shot classification on the **informative** task (2 classes: `informative`, `not_informative`) across all 3 modalities and both train/test splits.

| # | Modality | Split | Output |
|---|----------|-------|--------|
| 1 | text_only | test | `results/zeroshot/informative/text_only/test/` |
| 2 | text_only | train | `results/zeroshot/informative/text_only/train/` |
| 3 | image_only | test | `results/zeroshot/informative/image_only/test/` |
| 4 | image_only | train | `results/zeroshot/informative/image_only/train/` |
| 5 | text_image | test | `results/zeroshot/informative/text_image/test/` |
| 6 | text_image | train | `results/zeroshot/informative/text_image/train/` |

## Setup

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

from scripts.zeroshot_llama import load_model, run_zeroshot

# Configuration
MODEL_ID = "meta-llama/Llama-3.2-11B-Vision-Instruct"
TASK = "informative"
DATA_ROOT = os.path.abspath("../data")
OUTPUT_DIR = os.path.abspath("../results/zeroshot")

# Load model once — reused across all 6 experiments
model, processor = load_model(MODEL_ID)
```

## 1. Text-Only — Test Set

```python
preds_text_test, metrics_text_test = run_zeroshot(
    task=TASK, modality="text_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 2. Text-Only — Train Set

```python
preds_text_train, metrics_text_train = run_zeroshot(
    task=TASK, modality="text_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 3. Image-Only — Test Set

```python
preds_img_test, metrics_img_test = run_zeroshot(
    task=TASK, modality="image_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 4. Image-Only — Train Set

```python
preds_img_train, metrics_img_train = run_zeroshot(
    task=TASK, modality="image_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 5. Text+Image — Test Set

```python
preds_ti_test, metrics_ti_test = run_zeroshot(
    task=TASK, modality="text_image", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 6. Text+Image — Train Set

```python
preds_ti_train, metrics_ti_train = run_zeroshot(
    task=TASK, modality="text_image", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## Summary — All 6 Experiments

```python
import pandas as pd

all_metrics = [
    metrics_text_test, metrics_text_train,
    metrics_img_test, metrics_img_train,
    metrics_ti_test, metrics_ti_train,
]

summary = pd.DataFrame([
    {
        "Modality": m["modality"],
        "Split": m["split"],
        "Samples": m["num_samples"],
        "Accuracy": f"{m['accuracy']:.4f}",
        "Weighted F1": f"{m['weighted_f1']:.4f}",
        "Macro F1": f"{m['macro_f1']:.4f}",
        "Unparseable": m["num_unparseable"],
    }
    for m in all_metrics
])

print(f"Task: {TASK}")
print(f"Model: {MODEL_ID}\n")
display(summary)
```
