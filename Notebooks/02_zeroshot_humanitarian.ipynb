# 02 — Zero-Shot Humanitarian Classification (Llama-3.2-11B-Vision-Instruct)

Run zero-shot classification on the **humanitarian** task (5 classes) across all 3 modalities and both train/test splits.

**Classes:** `affected_individuals`, `infrastructure_and_utility_damage`, `not_humanitarian`, `other_relevant_information`, `rescue_volunteering_or_donation_effort`

| # | Modality | Split | Output |
|---|----------|-------|--------|
| 1 | text_only | test | `results/zeroshot/humanitarian/text_only/test/` |
| 2 | text_only | train | `results/zeroshot/humanitarian/text_only/train/` |
| 3 | image_only | test | `results/zeroshot/humanitarian/image_only/test/` |
| 4 | image_only | train | `results/zeroshot/humanitarian/image_only/train/` |
| 5 | text_image | test | `results/zeroshot/humanitarian/text_image/test/` |
| 6 | text_image | train | `results/zeroshot/humanitarian/text_image/train/` |

## Setup

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

from scripts.zeroshot_llama import load_model, run_zeroshot

# Configuration
MODEL_ID = "meta-llama/Llama-3.2-11B-Vision-Instruct"
TASK = "humanitarian"
DATA_ROOT = os.path.abspath("../data")
OUTPUT_DIR = os.path.abspath("../results/zeroshot")

# Load model once — reused across all 6 experiments
model, processor = load_model(MODEL_ID)
```

## 1. Text-Only — Test Set

```python
preds_text_test, metrics_text_test = run_zeroshot(
    task=TASK, modality="text_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 2. Text-Only — Train Set

```python
preds_text_train, metrics_text_train = run_zeroshot(
    task=TASK, modality="text_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 3. Image-Only — Test Set

```python
preds_img_test, metrics_img_test = run_zeroshot(
    task=TASK, modality="image_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 4. Image-Only — Train Set

```python
preds_img_train, metrics_img_train = run_zeroshot(
    task=TASK, modality="image_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 5. Text+Image — Test Set

```python
preds_ti_test, metrics_ti_test = run_zeroshot(
    task=TASK, modality="text_image", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## 6. Text+Image — Train Set

```python
preds_ti_train, metrics_ti_train = run_zeroshot(
    task=TASK, modality="text_image", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
```

## Summary — All 6 Experiments

```python
import pandas as pd

all_metrics = [
    metrics_text_test, metrics_text_train,
    metrics_img_test, metrics_img_train,
    metrics_ti_test, metrics_ti_train,
]

summary = pd.DataFrame([
    {
        "Modality": m["modality"],
        "Split": m["split"],
        "Samples": m["num_samples"],
        "Accuracy": f"{m['accuracy']:.4f}",
        "Weighted F1": f"{m['weighted_f1']:.4f}",
        "Macro F1": f"{m['macro_f1']:.4f}",
        "Unparseable": m["num_unparseable"],
    }
    for m in all_metrics
])

print(f"Task: {TASK}")
print(f"Model: {MODEL_ID}\n")
display(summary)

# Per-class F1 for test sets
print("\nPer-class F1 (test sets):")
for m in [metrics_text_test, metrics_img_test, metrics_ti_test]:
    print(f"\n  {m['modality']}:")
    for cls, f1 in sorted(m["per_class_f1"].items()):
        print(f"    {cls:<45} {f1:.4f}")
```

## Cleanup — Free GPU Memory

```python
import gc, torch

# Delete all large objects
del model, processor
del preds_text_test, preds_text_train
del preds_img_test, preds_img_train
del preds_ti_test, preds_ti_train

gc.collect()
torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()
print(f"GPU memory allocated: {torch.cuda.memory_allocated()/1e9:.1f} GB")
print(f"GPU memory reserved:  {torch.cuda.memory_reserved()/1e9:.1f} GB")
print("\nNote: If memory is still high, restart the kernel to fully release GPU memory.")
```
