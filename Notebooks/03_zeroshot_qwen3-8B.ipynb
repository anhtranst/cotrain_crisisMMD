# Zero-Shot Classification (Qwen3-VL-8B-Instruct)

Run zero-shot classification across **both tasks** and **all 3 modalities** (train + test splits).

**12 experiments total** (2 tasks × 3 modalities × 2 splits), run sequentially with cleanup between each.

| # | Task | Modality | Split | Output |
|---|------|----------|-------|--------|
| 1–6 | informative | text_only / image_only / text_image | test + train | `results/zeroshot/qwen3-vl-8b/informative/...` |
| 7–12 | humanitarian | text_only / image_only / text_image | test + train | `results/zeroshot/qwen3-vl-8b/humanitarian/...` |

## Setup

```python
import sys, os, gc
sys.path.insert(0, os.path.abspath(".."))

import torch
from scripts.zeroshot_qwen import load_model, run_zeroshot

# Configuration
MODEL_ID = "Qwen/Qwen3-VL-8B-Instruct"
DATA_ROOT = os.path.abspath("../data")
OUTPUT_DIR = os.path.abspath("../results/zeroshot")

def cleanup(*vars_to_del):
    """Delete prediction variables and free GPU memory."""
    for v in vars_to_del:
        del v
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

# Load model once -- reused across all 12 experiments
model, processor = load_model(MODEL_ID)
```

## Informative Task (6 experiments)

```python
# 1. Informative -- text_only -- test
preds, _ = run_zeroshot(
    task="informative", modality="text_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

```python
# 2. Informative -- text_only -- train
preds, _ = run_zeroshot(
    task="informative", modality="text_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

```python
# 3. Informative -- image_only -- test
preds, _ = run_zeroshot(
    task="informative", modality="image_only", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

```python
# 4. Informative -- image_only -- train
preds, _ = run_zeroshot(
    task="informative", modality="image_only", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

```python
# 5. Informative -- text_image -- test
preds, _ = run_zeroshot(
    task="informative", modality="text_image", split="test",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

```python
# 6. Informative -- text_image -- train
preds, _ = run_zeroshot(
    task="informative", modality="text_image", split="train",
    model_id=MODEL_ID, data_root=DATA_ROOT, output_dir=OUTPUT_DIR,
    model=model, processor=processor,
)
cleanup(preds)
```

### Informative Summary

```python
import json
import pandas as pd
from pathlib import Path
from scripts.zeroshot_qwen import model_slug

def show_summary(task):
    slug = model_slug(MODEL_ID)
    results_dir = Path(OUTPUT_DIR) / slug / task
    modalities = ["text_only", "image_only", "text_image"]
    splits = ["test", "train"]
    rows = []
    for modality in modalities:
        for split in splits:
            metrics_path = results_dir / modality / split / "metrics.json"
            if metrics_path.exists():
                m = json.loads(metrics_path.read_text())
                rows.append({
                    "Modality": m["modality"], "Split": m["split"],
                    "Samples": m["num_samples"],
                    "Accuracy": f"{m['accuracy']:.4f}",
                    "Weighted F1": f"{m['weighted_f1']:.4f}",
                    "Macro F1": f"{m['macro_f1']:.4f}",
                    "Unparseable": m["num_unparseable"],
                })
            else:
                rows.append({
                    "Modality": modality, "Split": split,
                    "Samples": "-", "Accuracy": "-",
                    "Weighted F1": "-", "Macro F1": "-", "Unparseable": "-",
                })
    print(f"Task: {task}")
    print(f"Model: {slug}\n")
    display(pd.DataFrame(rows))

show_summary("informative")
```
